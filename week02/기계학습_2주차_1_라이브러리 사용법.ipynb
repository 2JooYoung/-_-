### 2. 판다스 라이브러리 사용법

2.1 판다스 라이브러리 불러오기

```python
#판다스 라이브러리를 실행환경으로 불러오기 위해 import 명령을 사용한다.
# import 라이브러리명 as 라이브러리 약칭
import pandas as pd
```

2.2 딕셔너리(dictionary)를 시리즈(series)로 변환

```python
# {key:value} 구조를 갖는 딕셔너리를 만들고, 변수 dict_data에 저장한다.
dict_data = {'a':1,'b':2,'c':3}

#판다스 Series() 함수로 딕셔너리(dict_data)를 시리즈로 변환한다.
#Series() 함수가 반환한 시리즈 객체를 변수 sr에 저장한다.
sr = pd.Series(dict_data)

#변수 sr의 자료형 출력
print(type(sr))
print('\n')

#변수 sr에 저장되어 있는 시리즈 객체를 출력한다.
print(sr)
```

2.3 파이썬 리스트(list)를 시리즈(series)로 변환

```python
#파이썬 리스트는 어떠한 데이터형도 포함할 수 있다.
list_data = ['2019-01-02', 3.14, 'ABC', 100, True]

#판다스 Series() 함수로 리스트(list_data)를 시리즈로 변환한다.
#Sereis() 함수가 반환한 시리즈 객체를 변수 sr에 저장한다.
sr = pd.Series(list_data)
print(sr)
```

2.4 시리즈의 인덱스와 데이터 확인하기

```python
#시리즈의 index 속성은 인덱스 정보를 제공한다.
#인덱스 범위(Range Index)는 0부터 시작하고 마지막 인데스값(stop)은 포함되지 않는다.
print(sr.index)
print('\n')

#시리즈의 values 속성으로 데이터 배열을 불러온다.
print(sr.values)
```

2.5 파이썬 튜플(tuple)을 시리즈(series)로 변환

```python
# 튜플을 변수 tup_data에 저장한다.
tup_data = ('영인', '2010-05-01', '여', True)

# Series() 함수로 튜블(tup_data)를 시리즈로 변환한다.
# index 옵션에 인덱스 이름을 지정한다.
sr = pd.Series(tup_data, index=['이름', '생년월일', '성별', '학생여부'])
print(sr)
```

2.6 시리즈 원소 선택하기

```python
#원소를 1개 선택
print(sr[0])        #정수형 위치 인덱스를 활용
print(sr['이름'])   #인덱스 이름을 활용
```

```python
#여러 개의 원소를 선택 (인덱스 리스트 활용)
print(sr[[1,2]])
print('\n')
print(sr[['생년월일', '성별']])
```

```python
#여러 개의 원소를 선택 (인덱스 범위 지정)
# 정수형 인덱스 범위의 마지막 값은 포함되지 않는다.
print(sr[1 : 2])
print('\n')
# 인덱스 이름을 사용할 때는 마지막 값이 포함된다.
print(sr['생년월일':'성별'])
```

2.7 딕셔너리(dictionary)를 데이터프레임으로 변환

```python
# 열이름은 key로 하고, 리스트를 value로 갖는 딕셔너리를 정의한다.
dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}

# 판다스 DataFrame() 함수로 딕셔너리를 데이터프레임으로 변환한다.
# DataFrame() 함수가 변환한 데이터프레임 객체를 변수 df에 저장한다.
df = pd.DataFrame(dict_data)

#df의 자료형 출력한다.
print(type(df))
print('\n')
#변수 df에 저장되어 있는 데이터프레임 객체를 출력
print(df)
```

2.8 여러개의 리스트(list)를 데이터프레임으로 변환

```python
#먼저 개수가 같은 3개의 리스트를 생성한다.
names = ['Bob', 'Jessica', 'Mary', 'John', 'Mel']
births = [968, 155, 77, 578, 973]
custom = [1, 5, 25, 13, 23232]

#zip()는 길이가 같은 리스트 등의 요소를 묶어준다.
#list()는 리스트 자료형으로 변환해준다.
BabyDataSet = list(zip(names,births))

#pd.DataFrame(data=데이터셋, columns=컬럼명)으로 데이터프레임이 생성된다.
df = pd.DataFrame(data=BabyDataSet, columns=['Names', 'Births'])

#데이터프레임의 상위 5개의 행을 출력한다.
df.head()
```

2.9 데이터프레임의 기본 정보 출력하기

```python
#데이터프레임의 타입(type) 정보 --> 데이터 타입을 알 수 있다.
print(df.dtypes)
print("-----------")

#데이터프레임의 index 정보 --> 행의 수를 알 수 있다.
print(df.index)
print("-----------")

#데이터프레임의 열의 정보 --> 칼럼명을 알 수 있다.
print(df.columns)
print("-----------")
print(df.values)
```

2.10 데이터프레임의 특정 행ㅇ과 열 선택하기

```python
#데이터프레임에서 하나의 열을 선택한다.
df['Names']
```

2.11 데이터프레임의 인덱스 선택하기

```python
#데이터프레임의 특정 행(인덱스)를 선택한다.
#df[start:stop] start인덱스는 0부터 시작, (stop인덱스-1) 까지 행이 선택된다.
df[0:3]
```

2.12 데이터프레임의 필터링 기능

```python
#Births 열에서 100보다 큰 데이터를 필터링한다.
df[df['Births']>100]
```

2.13 평균값을 구하는 기능

```python
#mean()는 각 열의 데이터타입을 체크한 뒤, 수치형 데이터에 대해 평균을 구한다.
df.mean()
```
